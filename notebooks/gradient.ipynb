```python
import numpy as np
```

```python
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
```

```python
import matplotlib.pyplot as plt
```

```python
Z = X**2 + Y**2
plt.figure(figsize=(10, 10))
plt.contour(X, Y, Z)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
z = Z[:, Z.shape[1] // 2]
plt.plot(x, z)
plt.show()
```

```python
Z = X**2 + 20 * Y**2
plt.figure(figsize=(10, 10))
#plt.scatter([0], [0], color='blue')
#plt.scatter([2], [2], color='green')
plt.contour(X, Y, Z)
plt.show()
```

```python
def gradient(x, y): # f = x^2 + 20y^2
    return 2 * x, 40 * y    
```

```python
start = np.array([2, 2])
gradient(*start)
```

```python
next_point = start - 0.05 * np.array(gradient(*start))
```

```python
Z = X**2 + 20 * Y**2
plt.figure(figsize=(10, 10))
plt.scatter([0], [0], color='blue')
plt.scatter([2], [2], color='green')
plt.scatter([next_point[0]], [next_point[1]], color='orange')
plt.contour(X, Y, Z)
plt.show()
```

```python
points = [start]
for i in range(10):
    points.append(points[-1] - 0.05 * np.array(gradient(*points[-1])))

plt.figure(figsize=(10, 10))
plt.scatter([0], [0], color='blue')
plt.scatter([start[0]], [start[1]], color='green')
plt.scatter([x[0] for x in points[1:]], [x[1] for x in points[1:]], color='orange')
plt.plot([x[0] for x in points], [x[1] for x in points], color='orange')
plt.contour(X, Y, Z)
plt.show()
```

```python
# Momentum

points = [start]
grads = [[0, 0]]
for i in range(100):
    grads.append(gradient(*points[-1]) + 0.5 * np.array(grads[-1]))
    points.append(points[-1] - 0.05 * np.array(grads[-1]))

plt.figure(figsize=(10, 10))
plt.scatter([0], [0], color='blue')
plt.scatter([start[0]], [start[1]], color='green')
plt.scatter([x[0] for x in points[1:]], [x[1] for x in points[1:]], color='orange')
plt.plot([x[0] for x in points], [x[1] for x in points], color='orange')
plt.contour(X, Y, Z)
plt.show()
```
